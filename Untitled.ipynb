```python
# ============================================
# PART 1: PROCESS TAMIL DATASET
# ============================================

print("\n" + "="*60)
print("PART 1: PROCESSING TAMIL DATASET")
print("="*60)

def process_tamil(tamil_folder, output_base):
    """Process Tamil parquet files from nested folders"""
    
    tamil_audio_dir = os.path.join(output_base, "audio", "tamil")
    
    # Find all parquet files recursively
    tamil_files = glob.glob(os.path.join(tamil_folder, "**", "*.parquet"), recursive=True)
    print(f"Found {len(tamil_files)} Tamil parquet files")
    
    if len(tamil_files) == 0:
        print("❌ No Tamil files found!")
        return []
    
    all_tamil_data = []
    total_files = 0
    failed = 0
    
    for parquet_file in tqdm(tamil_files, desc="Processing Tamil"):
        try:
            df = pd.read_parquet(parquet_file)
            district = os.path.basename(os.path.dirname(parquet_file))
            
            for idx, row in df.iterrows():
                try:
                    audio_info = row['audio']
                    if 'bytes' in audio_info:
                        audio_bytes = audio_info['bytes']
                        
                        if 'path' in audio_info:
                            audio_path_string = audio_info['path']
                            filename = os.path.basename(audio_path_string)
                            parts = filename.split('_')
                            
                            if len(parts) > 5:
                                speaker_id = parts[5]
                            else:
                                speaker_id = f"unknown_{idx}"
                            
                            in_memory = io.BytesIO(audio_bytes)
                            audio_data, sr = sf.read(in_memory, dtype='float32')
                            
                            out_filename = f"tamil_spk_{speaker_id}_{district}_{idx:06d}.wav"
                            out_path = os.path.join(tamil_audio_dir, out_filename)
                            sf.write(out_path, audio_data, sr)
                            
                            duration = len(audio_data) / sr
                            
                            entry = {
                                'audio_path': out_path,
                                'speaker_id': f"tamil_{speaker_id}",
                                'language': 'tamil',
                                'district': district,
                                'duration': round(duration, 2)
                            }
                            all_tamil_data.append(entry)
                            total_files += 1
                except:
                    failed += 1
                    continue
        except:
            continue
    
    if all_tamil_data:
        manifest_path = os.path.join(output_base, "manifests", "tamil_all.jsonl")
        with open(manifest_path, 'w') as f:
            for entry in all_tamil_data:
                f.write(json.dumps(entry) + '\n')
        
        print(f"\n✅ Tamil Complete:")
        print(f"  Files: {total_files:,}")
        print(f"  Speakers: {len(set([e['speaker_id'] for e in all_tamil_data])):,}")
        print(f"  Hours: {sum([e['duration'] for e in all_tamil_data])/3600:.2f}")
        print(f"  Failed: {failed}")
    
    return all_tamil_data

# Call the function with defined variables
tamil_data = process_tamil(TAMIL_FOLDER, OUTPUT_BASE)
```

```python
# ============================================
# PART 4: CREATE TRAIN/TEST SPLITS
# ============================================

print("\n" + "="*60)
print("PART 4: CREATING TRAIN/TEST SPLITS")
print("="*60)

import json
import random
from collections import defaultdict
import os

OUTPUT_BASE = "/home/ladmin/new_env_nemo/my_local_data/processed_dataset"

def create_splits(output_base):
    """Create both split strategies"""
    
    # Load all data
    all_data = []
    for lang in ['tamil', 'telugu', 'english']:
        manifest = os.path.join(output_base, "manifests", f"{lang}_all.jsonl")
        if os.path.exists(manifest):
            with open(manifest, 'r') as f:
                for line in f:
                    all_data.append(json.loads(line))
            print(f"✅ Loaded {lang}: {len([d for d in all_data if d['language']==lang]):,} files")
    
    print(f"\n📊 Total loaded: {len(all_data):,} files")
    
    # Group by speaker
    speaker_to_files = defaultdict(list)
    for entry in all_data:
        speaker_to_files[entry['speaker_id']].append(entry)
    
    all_speakers = list(speaker_to_files.keys())
    print(f"👥 Total speakers: {len(all_speakers):,}")
    
    # ==========================================
    # SPLIT 1: Same speakers, different files (Identification)
    # ==========================================
    print("\n" + "-"*50)
    print("📁 SPLIT 1: Same Speakers, Different Files")
    print("-"*50)
    
    split1_train = []
    split1_val = []
    split1_test = []
    
    for speaker, files in speaker_to_files.items():
        random.shuffle(files)
        n = len(files)
        
        # 70% train, 15% val, 15% test
        train_idx = int(0.7 * n)
        val_idx = int(0.85 * n)
        
        split1_train.extend(files[:train_idx])
        split1_val.extend(files[train_idx:val_idx])
        split1_test.extend(files[val_idx:])
    
    print(f"Train files: {len(split1_train):,}")
    print(f"Val files: {len(split1_val):,}")
    print(f"Test files: {len(split1_test):,}")
    print(f"Train speakers: {len(set([f['speaker_id'] for f in split1_train])):,}")
    print(f"Test speakers: {len(set([f['speaker_id'] for f in split1_test])):,} ✓")
    
    # Save Split 1
    with open(os.path.join(output_base, "manifests", "split1_train.jsonl"), 'w') as f:
        for e in split1_train:
            f.write(json.dumps(e) + '\n')
    
    with open(os.path.join(output_base, "manifests", "split1_val.jsonl"), 'w') as f:
        for e in split1_val:
            f.write(json.dumps(e) + '\n')
    
    with open(os.path.join(output_base, "manifests", "split1_test.jsonl"), 'w') as f:
        for e in split1_test:
            f.write(json.dumps(e) + '\n')
    
    print("✅ Split 1 saved")
    
    # ==========================================
    # SPLIT 2: Different speakers (Open-set)
    # ==========================================
    print("\n" + "-"*50)
    print("📁 SPLIT 2: Different Speakers (Open-set)")
    print("-"*50)
    
    random.shuffle(all_speakers)
    n_speakers = len(all_speakers)
    
    train_spk = set(all_speakers[:int(0.8 * n_speakers)])
    val_spk = set(all_speakers[int(0.8 * n_speakers):int(0.9 * n_speakers)])
    test_spk = set(all_speakers[int(0.9 * n_speakers):])
    
    split2_train = []
    split2_val = []
    split2_test = []
    
    for speaker, files in speaker_to_files.items():
        if speaker in train_spk:
            split2_train.extend(files)
        elif speaker in val_spk:
            split2_val.extend(files)
        else:
            split2_test.extend(files)
    
    print(f"Train speakers: {len(train_spk):,}")
    print(f"Val speakers: {len(val_spk):,}")
    print(f"Test speakers: {len(test_spk):,}")
    print(f"Train files: {len(split2_train):,}")
    print(f"Val files: {len(split2_val):,}")
    print(f"Test files: {len(split2_test):,}")
    
    # Save Split 2
    with open(os.path.join(output_base, "manifests", "split2_train.jsonl"), 'w') as f:
        for e in split2_train:
            f.write(json.dumps(e) + '\n')
    
    with open(os.path.join(output_base, "manifests", "split2_val.jsonl"), 'w') as f:
        for e in split2_val:
            f.write(json.dumps(e) + '\n')
    
    with open(os.path.join(output_base, "manifests", "split2_test.jsonl"), 'w') as f:
        for e in split2_test:
            f.write(json.dumps(e) + '\n')
    
    print("✅ Split 2 saved")
    
    return {
        'split1': {
            'train': len(split1_train),
            'val': len(split1_val),
            'test': len(split1_test)
        },
        'split2': {
            'train': len(split2_train),
            'val': len(split2_val),
            'test': len(split2_test)
        }
    }

# Run the split creation
split_stats = create_splits(OUTPUT_BASE)
```

```python
# ============================================
# PART 5: GENERATE STATISTICS FOR PAPER
# ============================================

print("\n" + "="*60)
print("PART 5: GENERATING PAPER STATISTICS")
print("="*60)

def generate_paper_stats(output_base):
    """Generate all statistics needed for paper"""
    
    # Load all data
    all_data = []
    for lang in ['tamil', 'telugu', 'english']:
        manifest = os.path.join(output_base, "manifests", f"{lang}_all.jsonl")
        if os.path.exists(manifest):
            with open(manifest, 'r') as f:
                for line in f:
                    all_data.append(json.loads(line))
    
    # Per language stats
    stats = {}
    for lang in ['tamil', 'telugu', 'english']:
        lang_data = [d for d in all_data if d['language'] == lang]
        if lang_data:
            speakers = set([d['speaker_id'] for d in lang_data])
            hours = sum([d['duration'] for d in lang_data]) / 3600
            
            stats[lang] = {
                'files': len(lang_data),
                'speakers': len(speakers),
                'hours': round(hours, 2),
                'avg_duration': round(sum([d['duration'] for d in lang_data]) / len(lang_data), 2),
                'files_per_speaker': round(len(lang_data) / len(speakers), 1)
            }
    
    # Total stats
    all_speakers = set([d['speaker_id'] for d in all_data])
    stats['total'] = {
        'files': len(all_data),
        'speakers': len(all_speakers),
        'hours': round(sum([d['duration'] for d in all_data]) / 3600, 2),
        'avg_duration': round(sum([d['duration'] for d in all_data]) / len(all_data), 2)
    }
    
    # Save stats
    with open(os.path.join(output_base, "stats", "paper_stats.json"), 'w') as f:
        json.dump(stats, f, indent=2)
    
    # Print beautiful table
    print("\n" + "="*60)
    print("📊 DATASET STATISTICS FOR IEEE PAPER")
    print("="*60)
    print("\n{:<12} {:>12} {:>12} {:>12} {:>12} {:>12}".format(
        "Language", "Files", "Speakers", "Hours", "Avg(s)", "Files/Spk"))
    print("-"*80)
    
    for lang in ['tamil', 'telugu', 'english', 'total']:
        if lang in stats:
            if lang == 'total':
                print("-"*80)
                label = "TOTAL"
            else:
                label = lang.capitalize()
            
            if lang == 'total':
                print("{:<12} {:>12,} {:>12,} {:>12.2f} {:>12.2f}".format(
                    label,
                    stats[lang]['files'],
                    stats[lang]['speakers'],
                    stats[lang]['hours'],
                    stats[lang]['avg_duration']
                ))
            else:
                print("{:<12} {:>12,} {:>12,} {:>12.2f} {:>12.2f} {:>12.1f}".format(
                    label,
                    stats[lang]['files'],
                    stats[lang]['speakers'],
                    stats[lang]['hours'],
                    stats[lang]['avg_duration'],
                    stats[lang]['files_per_speaker']
                ))
    
    # Generate LaTeX table for paper
    latex = r"""
\begin{table}[h]
\centering
\caption{Summary of the multilingual speaker identification dataset comprising Tamil, Telugu, and Indian English.}
\label{tab:dataset}
\begin{tabular}{|l|c|c|c|c|c|}
\hline
\textbf{Language} & \textbf{Speakers} & \textbf{Utterances} & \textbf{Hours} & \textbf{Avg. Dur. (s)} & \textbf{Utterances/Speaker} \\
\hline"""
    
    for lang in ['tamil', 'telugu', 'english']:
        if lang in stats:
            latex += f"\n{lang.capitalize()} & {stats[lang]['speakers']:,} & {stats[lang]['files']:,} & {stats[lang]['hours']:.1f} & {stats[lang]['avg_duration']:.2f} & {stats[lang]['files_per_speaker']:.1f} \\\\"
    
    latex += r"""
\hline
\textbf{Total} & """ + f"{stats['total']['speakers']:,} & {stats['total']['files']:,} & {stats['total']['hours']:.1f} & {stats['total']['avg_duration']:.2f} & {stats['total']['files']/stats['total']['speakers']:.1f} \\\\" + r"""
\hline
\end{tabular}
\end{table}"""
    
    # Save LaTeX table
    with open(os.path.join(output_base, "stats", "dataset_table.tex"), 'w') as f:
        f.write(latex)
    
    print(f"\n✅ LaTeX table saved to: stats/dataset_table.tex")
    
    # Create visualization
    try:
        import matplotlib.pyplot as plt
        
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        
        # Files
        langs = ['Tamil', 'Telugu', 'English']
        files = [stats['tamil']['files'], stats['telugu']['files'], stats['english']['files']]
        colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
        
        axes[0,0].bar(langs, files, color=colors)
        axes[0,0].set_title('Number of Audio Files', fontsize=14, fontweight='bold')
        axes[0,0].set_ylabel('Files', fontsize=12)
        for i, v in enumerate(files):
            axes[0,0].text(i, v + 10000, f'{v:,}', ha='center', fontsize=10)
        
        # Speakers
        speakers = [stats['tamil']['speakers'], stats['telugu']['speakers'], stats['english']['speakers']]
        axes[0,1].bar(langs, speakers, color=colors)
        axes[0,1].set_title('Number of Speakers', fontsize=14, fontweight='bold')
        axes[0,1].set_ylabel('Speakers', fontsize=12)
        for i, v in enumerate(speakers):
            axes[0,1].text(i, v + 100, f'{v:,}', ha='center', fontsize=10)
        
        # Hours
        hours = [stats['tamil']['hours'], stats['telugu']['hours'], stats['english']['hours']]
        axes[1,0].bar(langs, hours, color=colors)
        axes[1,0].set_title('Total Duration (Hours)', fontsize=14, fontweight='bold')
        axes[1,0].set_ylabel('Hours', fontsize=12)
        for i, v in enumerate(hours):
            axes[1,0].text(i, v + 10, f'{v:.1f}', ha='center', fontsize=10)
        
        # Pie chart
        axes[1,1].pie(hours, labels=langs, autopct='%1.1f%%', colors=colors, startangle=90)
        axes[1,1].set_title('Dataset Distribution by Hours', fontsize=14, fontweight='bold')
        
        plt.tight_layout()
        plt.savefig(os.path.join(output_base, "stats", "dataset_stats.png"), dpi=300, bbox_inches='tight')
        plt.show()
        print(f"✅ Visualization saved to: stats/dataset_stats.png")
    except:
        print("⚠️ Matplotlib not available - skipping visualization")
    
    return stats

# Generate statistics
paper_stats = generate_paper_stats(OUTPUT_BASE)
```

```python
# ============================================
# FIXED: RECREATE SPLIT 1 - SAFE VERSION
# ============================================

print("\n" + "="*60)
print("🔄 RECREATING SPLIT 1 - SAFE VERSION")
print("="*60)

import json
import random
from collections import defaultdict
import os

OUTPUT_BASE = "/home/ladmin/new_env_nemo/my_local_data/processed_dataset"

# Load all data
all_data = []
for lang in ['tamil', 'telugu', 'english']:
    manifest = os.path.join(OUTPUT_BASE, "manifests", f"{lang}_all.jsonl")
    if os.path.exists(manifest):
        with open(manifest, 'r') as f:
            for line in f:
                all_data.append(json.loads(line))

print(f"📊 Total loaded: {len(all_data):,} files")

# Group by speaker
speaker_to_files = defaultdict(list)
for entry in all_data:
    speaker_to_files[entry['speaker_id']].append(entry)

# Use a fixed random seed
random.seed(42)

split1_train = []
split1_val = []
split1_test = []

# Track all assigned file paths to prevent ANY overlap
all_assigned_paths = set()

speakers_with_issues = 0

for speaker, files in speaker_to_files.items():
    # Get unique file paths for this speaker
    # Use dictionary to deduplicate by audio_path
    unique_files = {}
    for f in files:
        if f['audio_path'] not in unique_files:
            unique_files[f['audio_path']] = f
    
    files_list = list(unique_files.values())
    n = len(files_list)
    
    if n < 3:
        print(f"⚠️ Speaker {speaker} has only {n} files - skipping")
        continue
    
    # Shuffle
    random.shuffle(files_list)
    
    # Calculate split indices (ensure at least 1 file in each set if possible)
    train_idx = max(1, int(0.7 * n))
    val_idx = min(n - 1, int(0.85 * n))
    
    if train_idx >= val_idx:
        train_idx = max(1, n - 2)
        val_idx = n - 1
    
    # Split
    speaker_train = files_list[:train_idx]
    speaker_val = files_list[train_idx:val_idx]
    speaker_test = files_list[val_idx:]
    
    # Verify no overlap within this speaker using paths
    train_paths = set([f['audio_path'] for f in speaker_train])
    val_paths = set([f['audio_path'] for f in speaker_val])
    test_paths = set([f['audio_path'] for f in speaker_test])
    
    # Check internal overlaps
    if (len(train_paths.intersection(val_paths)) > 0 or 
        len(train_paths.intersection(test_paths)) > 0 or 
        len(val_paths.intersection(test_paths)) > 0):
        speakers_with_issues += 1
        print(f"⚠️ Speaker {speaker} has internal overlap - using alternative split")
        
        # Alternative: just use first n-2 for train, last 1 for val, last 1 for test
        if n >= 3:
            speaker_train = files_list[:n-2]
            speaker_val = files_list[n-2:n-1]
            speaker_test = files_list[n-1:]
    
    # Add to global sets
    split1_train.extend(speaker_train)
    split1_val.extend(speaker_val)
    split1_test.extend(speaker_test)
    
    # Track global assignments
    for f in speaker_train + speaker_val + speaker_test:
        all_assigned_paths.add(f['audio_path'])

print(f"\n📁 Split 1 Statistics:")
print(f"Train files: {len(split1_train):,}")
print(f"Val files: {len(split1_val):,}")
print(f"Test files: {len(split1_test):,}")
print(f"Total unique files: {len(all_assigned_paths):,}")
print(f"Speakers with issues: {speakers_with_issues}")

# Final verification
train_paths = set([f['audio_path'] for f in split1_train])
val_paths = set([f['audio_path'] for f in split1_val])
test_paths = set([f['audio_path'] for f in split1_test])

train_val_overlap = len(train_paths.intersection(val_paths))
train_test_overlap = len(train_paths.intersection(test_paths))
val_test_overlap = len(val_paths.intersection(test_paths))

print(f"\n🔍 Final Verification:")
print(f"Train-Val overlap: {train_val_overlap} files")
print(f"Train-Test overlap: {train_test_overlap} files")
print(f"Val-Test overlap: {val_test_overlap} files")

if train_val_overlap == 0 and train_test_overlap == 0 and val_test_overlap == 0:
    print("✅ VERIFICATION PASSED: No overlap!")
    
    # Save corrected Split 1
    with open(os.path.join(OUTPUT_BASE, "manifests", "split1_train.jsonl"), 'w') as f:
        for e in split1_train:
            f.write(json.dumps(e) + '\n')
    
    with open(os.path.join(OUTPUT_BASE, "manifests", "split1_val.jsonl"), 'w') as f:
        for e in split1_val:
            f.write(json.dumps(e) + '\n')
    
    with open(os.path.join(OUTPUT_BASE, "manifests", "split1_test.jsonl"), 'w') as f:
        for e in split1_test:
            f.write(json.dumps(e) + '\n')
    
    print("\n✅ Split 1 saved successfully with NO overlap!")
    
    # Show sample counts
    print(f"\n📊 Final Split 1 composition:")
    print(f"  Training:   {len(split1_train):,} files")
    print(f"  Validation: {len(split1_val):,} files")
    print(f"  Test:       {len(split1_test):,} files")
    
else:
    print("\n❌ ERROR: Overlap still present!")
    print(f"Train-Val overlap: {train_val_overlap}")
    print(f"Train-Test overlap: {train_test_overlap}")
    print(f"Val-Test overlap: {val_test_overlap}")
```
