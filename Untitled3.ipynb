```python
# ============================================
# create_subset_manifests.py
# Run this to create smaller dataset versions
# ============================================

import json
import random
from collections import defaultdict
import os

# Configuration
ORIGINAL_MANIFESTS_DIR = "/home/ladmin/new_env_nemo/my_local_data/processed_dataset/manifests"
OUTPUT_DIR = "/home/ladmin/new_env_nemo/my_local_data/processed_dataset/manifests_subsets"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ============================================
# 1. CREATE 5% TEST VERSION (For quick testing)
# ============================================

print("\n" + "="*60)
print("📊 CREATING 5% TEST VERSION")
print("="*60)

def create_subset(manifest_path, percentage, output_name):
    """Create a subset of a manifest file"""
    
    with open(manifest_path, 'r') as f:
        lines = f.readlines()
    
    # Randomly sample percentage of lines
    num_samples = int(len(lines) * percentage / 100)
    sampled_lines = random.sample(lines, num_samples)
    
    # Save subset
    output_path = os.path.join(OUTPUT_DIR, output_name)
    with open(output_path, 'w') as f:
        for line in sampled_lines:
            f.write(line)
    
    print(f"✅ Created: {output_name}")
    print(f"   Original: {len(lines):,} files")
    print(f"   Subset: {len(sampled_lines):,} files ({percentage}%)")
    
    return len(sampled_lines)

# Create 5% versions of all split files
test_sizes = {}

for split in ['split1_train', 'split1_val', 'split1_test', 
              'split2_train', 'split2_val', 'split2_test']:
    manifest = os.path.join(ORIGINAL_MANIFESTS_DIR, f"{split}.jsonl")
    if os.path.exists(manifest):
        size = create_subset(manifest, 5, f"5percent_{split}.jsonl")
        test_sizes[split] = size

print("\n📈 5% Test Dataset Summary:")
for split, size in test_sizes.items():
    print(f"  {split}: {size:,} files")

# ============================================
# 2. CREATE 50% FULL DATASET VERSION
# ============================================

print("\n" + "="*60)
print("📊 CREATING 50% FULL DATASET VERSION")
print("="*60)

half_sizes = {}

for split in ['split1_train', 'split1_val', 'split1_test', 
              'split2_train', 'split2_val', 'split2_test']:
    manifest = os.path.join(ORIGINAL_MANIFESTS_DIR, f"{split}.jsonl")
    if os.path.exists(manifest):
        size = create_subset(manifest, 50, f"50percent_{split}.jsonl")
        half_sizes[split] = size

print("\n📈 50% Dataset Summary:")
for split, size in half_sizes.items():
    print(f"  {split}: {size:,} files")

# ============================================
# 3. VERIFY SPEAKER DISTRIBUTION (Optional)
# ============================================

def check_speaker_distribution(manifest_path):
    """Check if speaker distribution is preserved"""
    
    speakers = defaultdict(int)
    
    with open(manifest_path, 'r') as f:
        for line in f:
            data = json.loads(line)
            speakers[data['speaker_id']] += 1
    
    print(f"\n📊 Speaker distribution for {os.path.basename(manifest_path)}:")
    print(f"   Total speakers: {len(speakers)}")
    print(f"   Avg files/speaker: {sum(speakers.values())/len(speakers):.1f}")
    print(f"   Min files: {min(speakers.values())}")
    print(f"   Max files: {max(speakers.values())}")
    
    return speakers

# Check one file to verify
check_speaker_distribution(os.path.join(OUTPUT_DIR, "5percent_split1_train.jsonl"))

print("\n✅ Subset creation complete!")
print(f"\n📁 Files saved in: {OUTPUT_DIR}")
print("\nTo use these subsets, update your config.py paths:")
print('''
In config.py, change:
    SPLIT1_TRAIN = os.path.join(DATA_DIR, "manifests_subsets", "5percent_split1_train.jsonl")
    SPLIT1_VAL = os.path.join(DATA_DIR, "manifests_subsets", "5percent_split1_val.jsonl")
    SPLIT1_TEST = os.path.join(DATA_DIR, "manifests_subsets", "5percent_split1_test.jsonl")
''')
```

```python
# ============================================
# create_stratified_subsets.py
# Creates subsets preserving language distribution
# ============================================

import json
import random
from collections import defaultdict
import os

# Configuration
ORIGINAL_MANIFESTS_DIR = "/home/ladmin/new_env_nemo/my_local_data/processed_dataset/manifests"
OUTPUT_DIR = "/home/ladmin/new_env_nemo/my_local_data/processed_dataset/manifests_subsets"
os.makedirs(OUTPUT_DIR, exist_ok=True)

def create_stratified_subset(manifest_path, language_percentages, output_name):
    """
    Create subset with different percentages per language
    
    Args:
        manifest_path: Path to original manifest
        language_percentages: Dict like {'tamil': 5, 'telugu': 5, 'english': 5}
        output_name: Output filename
    """
    
    # Group files by language
    language_files = defaultdict(list)
    
    with open(manifest_path, 'r') as f:
        for line in f:
            data = json.loads(line)
            lang = data['language']
            language_files[lang].append(data)
    
    print(f"\n📊 Original distribution in {os.path.basename(manifest_path)}:")
    sampled_files = []
    
    for lang, files in language_files.items():
        print(f"   {lang}: {len(files):,} files")
        
        # Get percentage for this language
        percentage = language_percentages.get(lang, 0)
        
        if percentage == 100:
            # Take all files
            sampled = files
            print(f"     → Taking 100%: {len(files):,} files")
        else:
            # Sample percentage
            num_samples = int(len(files) * percentage / 100)
            sampled = random.sample(files, num_samples)
            print(f"     → Taking {percentage}%: {num_samples:,} files")
        
        sampled_files.extend(sampled)
    
    # Shuffle mixed files
    random.shuffle(sampled_files)
    
    # Save subset
    output_path = os.path.join(OUTPUT_DIR, output_name)
    with open(output_path, 'w') as f:
        for item in sampled_files:
            f.write(json.dumps(item) + '\n')
    
    print(f"\n✅ Created: {output_name}")
    print(f"   Total files: {len(sampled_files):,}")
    
    # Show final distribution
    final_dist = defaultdict(int)
    for item in sampled_files:
        final_dist[item['language']] += 1
    
    print("   Final distribution:")
    for lang, count in final_dist.items():
        print(f"     {lang}: {count:,} files")
    
    return len(sampled_files)

# ============================================
# 1. CREATE 5% TEST VERSION (5% each language)
# ============================================

print("\n" + "="*60)
print("📊 CREATING 5% TEST VERSION (5% Tamil, 5% Telugu, 30% English)")
print("="*60)

five_percent_config = {
    'tamil': 5,
    'telugu': 5,
    'english': 30
}

for split in ['split1_train', 'split1_val', 'split1_test', 
              'split2_train', 'split2_val', 'split2_test']:
    manifest = os.path.join(ORIGINAL_MANIFESTS_DIR, f"{split}.jsonl")
    if os.path.exists(manifest):
        create_stratified_subset(
            manifest, 
            five_percent_config,
            f"5percent_{split}.jsonl"
        )

# ============================================
# 2. CREATE 50% VERSION (50% Tamil, 50% Telugu, 100% English)
# ============================================

print("\n" + "="*60)
print("📊 CREATING 50% VERSION (50% Tamil, 50% Telugu, 100% English)")
print("="*60)

fifty_percent_config = {
    'tamil': 50,
    'telugu': 50,
    'english': 100  # Take all English files
}

for split in ['split1_train', 'split1_val', 'split1_test', 
              'split2_train', 'split2_val', 'split2_test']:
    manifest = os.path.join(ORIGINAL_MANIFESTS_DIR, f"{split}.jsonl")
    if os.path.exists(manifest):
        create_stratified_subset(
            manifest, 
            fifty_percent_config,
            f"50percent_{split}.jsonl"
        )

# ============================================
# 3. CREATE SUMMARY REPORT
# ============================================

print("\n" + "="*60)
print("📈 SUBSET SUMMARY")
print("="*60)

original_totals = defaultdict(lambda: defaultdict(int))

# Calculate original totals
for split in ['split1_train', 'split1_val', 'split1_test']:
    manifest = os.path.join(ORIGINAL_MANIFESTS_DIR, f"{split}.jsonl")
    if os.path.exists(manifest):
        with open(manifest, 'r') as f:
            for line in f:
                data = json.loads(line)
                original_totals[split][data['language']] += 1

print("\n📊 Original Split 1 Totals:")
for split in ['split1_train', 'split1_val', 'split1_test']:
    print(f"\n{split}:")
    for lang in ['tamil', 'telugu', 'english']:
        count = original_totals[split].get(lang, 0)
        print(f"  {lang}: {count:,} files")

print("\n✅ Subset creation complete!")
print(f"\n📁 Files saved in: {OUTPUT_DIR}")
print("\n📝 Files created:")
for file in os.listdir(OUTPUT_DIR):
    if file.endswith('.jsonl'):
        size = os.path.getsize(os.path.join(OUTPUT_DIR, file)) / (1024**2)
        print(f"  {file}: {size:.1f} MB")
```
