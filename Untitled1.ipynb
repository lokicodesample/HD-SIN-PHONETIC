```python
# ============================================
# BASELINE ARCHITECTURES FOR SPEAKER IDENTIFICATION
# A: Classifier Fine-tuning
# B: PEFT Metric Learning  
# C: Full Model Fine-tuning
# ============================================

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from transformers import WavLMModel, WavLMConfig
import json
import os
import soundfile as sf
import numpy as np
from tqdm import tqdm
import random
from collections import defaultdict

# ============================================
# CONFIGURATION
# ============================================

class Config:
    # Paths
    DATA_DIR = "/home/ladmin/new_env_nemo/my_local_data/processed_dataset"
    SPLIT1_TRAIN = os.path.join(DATA_DIR, "manifests", "split1_train.jsonl")
    SPLIT1_VAL = os.path.join(DATA_DIR, "manifests", "split1_val.jsonl")
    SPLIT1_TEST = os.path.join(DATA_DIR, "manifests", "split1_test.jsonl")
    
    # Audio parameters
    SAMPLE_RATE = 16000
    MAX_DURATION = 8.0  # seconds
    MAX_LENGTH = int(MAX_DURATION * SAMPLE_RATE)
    
    # Training parameters
    BATCH_SIZE = 8
    LEARNING_RATE = 1e-4
    EPOCHS = 20
    DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
    # Model parameters
    NUM_SPEAKERS = 11315  # Total speakers
    EMBEDDING_DIM = 256

config = Config()
print(f"✅ Configuration loaded")
print(f"   Device: {config.DEVICE}")
print(f"   Num speakers: {config.NUM_SPEAKERS}")
```

```python
# ============================================
# DATASET CLASS
# ============================================

class SpeakerDataset(Dataset):
    """Dataset for speaker identification"""
    
    def __init__(self, manifest_path, config, is_training=True):
        self.config = config
        self.is_training = is_training
        
        # Load manifest
        self.data = []
        with open(manifest_path, 'r') as f:
            for line in f:
                self.data.append(json.loads(line))
        
        # Create speaker to index mapping
        self.speaker_to_idx = {}
        self.idx_to_speaker = []
        for item in self.data:
            if item['speaker_id'] not in self.speaker_to_idx:
                idx = len(self.speaker_to_idx)
                self.speaker_to_idx[item['speaker_id']] = idx
                self.idx_to_speaker.append(item['speaker_id'])
        
        print(f"📊 Dataset loaded: {len(self.data)} files, {len(self.speaker_to_idx)} speakers")
    
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        item = self.data[idx]
        
        # Load audio
        audio_path = item['audio_path']
        audio, sr = sf.read(audio_path)
        
        # Resample if needed (should already be 16kHz)
        if sr != self.config.SAMPLE_RATE:
            import librosa
            audio = librosa.resample(audio, orig_sr=sr, target_sr=self.config.SAMPLE_RATE)
        
        # Pad or truncate to fixed length
        if len(audio) > self.config.MAX_LENGTH:
            audio = audio[:self.config.MAX_LENGTH]
        else:
            padding = self.config.MAX_LENGTH - len(audio)
            audio = np.pad(audio, (0, padding), 'constant')
        
        # Convert to tensor
        audio = torch.FloatTensor(audio)
        
        # Get speaker label
        speaker_id = item['speaker_id']
        label = self.speaker_to_idx[speaker_id]
        
        return {
            'audio': audio,
            'label': label,
            'speaker_id': speaker_id,
            'path': audio_path
        }


class TripletDataset(Dataset):
    """Dataset for triplet loss training"""
    
    def __init__(self, manifest_path, config):
        self.config = config
        
        # Load manifest
        self.data = []
        with open(manifest_path, 'r') as f:
            for line in f:
                self.data.append(json.loads(line))
        
        # Group by speaker
        self.speaker_to_files = defaultdict(list)
        for item in self.data:
            self.speaker_to_files[item['speaker_id']].append(item)
        
        self.speakers = list(self.speaker_to_files.keys())
        print(f"📊 Triplet dataset: {len(self.data)} files, {len(self.speakers)} speakers")
    
    def __len__(self):
        return len(self.data) // 3  # Each triplet uses 3 files
    
    def __getitem__(self, idx):
        # Select anchor speaker
        anchor_speaker = random.choice(self.speakers)
        anchor_files = self.speaker_to_files[anchor_speaker]
        
        # Select anchor and positive from same speaker
        anchor_item = random.choice(anchor_files)
        positive_item = random.choice([f for f in anchor_files if f != anchor_item])
        
        # Select negative from different speaker
        negative_speaker = random.choice([s for s in self.speakers if s != anchor_speaker])
        negative_item = random.choice(self.speaker_to_files[negative_speaker])
        
        # Load and process audio
        def load_audio(item):
            audio, sr = sf.read(item['audio_path'])
            if sr != config.SAMPLE_RATE:
                import librosa
                audio = librosa.resample(audio, orig_sr=sr, target_sr=config.SAMPLE_RATE)
            if len(audio) > config.MAX_LENGTH:
                audio = audio[:config.MAX_LENGTH]
            else:
                padding = config.MAX_LENGTH - len(audio)
                audio = np.pad(audio, (0, padding), 'constant')
            return torch.FloatTensor(audio)
        
        anchor_audio = load_audio(anchor_item)
        positive_audio = load_audio(positive_item)
        negative_audio = load_audio(negative_item)
        
        return {
            'anchor': anchor_audio,
            'positive': positive_audio,
            'negative': negative_audio,
            'anchor_speaker': anchor_speaker,
            'negative_speaker': negative_speaker
        }
```

```python
# ============================================
# ARCHITECTURE A: Classifier Fine-tuning
# Frozen WavLM + Trainable Classifier
# ============================================

class WavLMClassifier(nn.Module):
    """Frozen WavLM with trainable classifier head"""
    
    def __init__(self, config):
        super().__init__()
        
        print("🔄 Loading WavLM-large (frozen)...")
        self.wavlm = WavLMModel.from_pretrained("microsoft/wavlm-large")
        self.wavlm.eval()  # Set to eval mode
        
        # Freeze all WavLM parameters
        for param in self.wavlm.parameters():
            param.requires_grad = False
        
        # Get WavLM hidden size
        wavlm_config = WavLMConfig.from_pretrained("microsoft/wavlm-large")
        hidden_size = wavlm_config.hidden_size  # Should be 1024
        
        # Weighted layer pooling (trainable)
        self.layer_weights = nn.Parameter(torch.ones(24) / 24)  # 24 layers
        
        # Classifier head (trainable)
        self.classifier = nn.Sequential(
            nn.Linear(hidden_size, 512),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(512, config.NUM_SPEAKERS)
        )
        
        print(f"✅ Architecture A initialized")
        print(f"   Trainable params: {sum(p.numel() for p in self.parameters() if p.requires_grad):,}")
    
    def forward(self, audio):
        # audio: [batch_size, max_length]
        
        # Get WavLM outputs with hidden states
        outputs = self.wavlm(audio, output_hidden_states=True)
        hidden_states = outputs.hidden_states  # 25 layers (embedding + 24 transformer)
        
        # Weighted sum of all layers
        weighted_sum = torch.zeros_like(hidden_states[-1])
        for i, hidden in enumerate(hidden_states[1:]):  # Skip embedding layer
            weighted_sum += self.layer_weights[i] * hidden
        
        # Mean pooling over time
        pooled = weighted_sum.mean(dim=1)  # [batch_size, hidden_size]
        
        # Classify
        logits = self.classifier(pooled)
        
        return logits, pooled


def train_architecture_a(train_loader, val_loader, config):
    """Training loop for Architecture A"""
    
    model = WavLMClassifier(config).to(config.DEVICE)
    optimizer = torch.optim.Adam(model.parameters(), lr=config.LEARNING_RATE)
    criterion = nn.CrossEntropyLoss()
    
    print("\n🚀 Training Architecture A (Classifier)")
    print("="*60)
    
    best_acc = 0
    
    for epoch in range(config.EPOCHS):
        # Training
        model.train()
        train_loss = 0
        train_correct = 0
        train_total = 0
        
        for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{config.EPOCHS} [Train]"):
            audio = batch['audio'].to(config.DEVICE)
            labels = batch['label'].to(config.DEVICE)
            
            optimizer.zero_grad()
            logits, _ = model(audio)
            loss = criterion(logits, labels)
            
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            _, predicted = logits.max(1)
            train_total += labels.size(0)
            train_correct += predicted.eq(labels).sum().item()
        
        train_acc = 100. * train_correct / train_total
        
        # Validation
        model.eval()
        val_loss = 0
        val_correct = 0
        val_total = 0
        
        with torch.no_grad():
            for batch in tqdm(val_loader, desc=f"Epoch {epoch+1}/{config.EPOCHS} [Val]"):
                audio = batch['audio'].to(config.DEVICE)
                labels = batch['label'].to(config.DEVICE)
                
                logits, _ = model(audio)
                loss = criterion(logits, labels)
                
                val_loss += loss.item()
                _, predicted = logits.max(1)
                val_total += labels.size(0)
                val_correct += predicted.eq(labels).sum().item()
        
        val_acc = 100. * val_correct / val_total
        
        print(f"\nEpoch {epoch+1}: Train Acc: {train_acc:.2f}%, Val Acc: {val_acc:.2f}%")
        
        # Save best model
        if val_acc > best_acc:
            best_acc = val_acc
            torch.save(model.state_dict(), "architecture_a_best.pt")
            print(f"✅ Saved best model with val acc: {val_acc:.2f}%")
    
    return model
```

```python
# ============================================
# ARCHITECTURE B: PEFT Metric Learning
# Frozen WavLM + Trainable Pooling + Triplet Loss
# ============================================

class WavLMTriplet(nn.Module):
    """Frozen WavLM with trainable pooling for triplet loss"""
    
    def __init__(self, config):
        super().__init__()
        
        print("🔄 Loading WavLM-large (frozen)...")
        self.wavlm = WavLMModel.from_pretrained("microsoft/wavlm-large")
        self.wavlm.eval()
        
        # Freeze all WavLM parameters
        for param in self.wavlm.parameters():
            param.requires_grad = False
        
        # Get WavLM hidden size
        wavlm_config = WavLMConfig.from_pretrained("microsoft/wavlm-large")
        hidden_size = wavlm_config.hidden_size  # 1024
        
        # Trainable components
        self.layer_weights = nn.Parameter(torch.ones(24) / 24)
        self.embedding_proj = nn.Linear(hidden_size, config.EMBEDDING_DIM)
        
        print(f"✅ Architecture B initialized")
        print(f"   Trainable params: {sum(p.numel() for p in self.parameters() if p.requires_grad):,}")
    
    def forward(self, audio):
        # audio: [batch_size, max_length]
        
        # Get WavLM outputs
        outputs = self.wavlm(audio, output_hidden_states=True)
        hidden_states = outputs.hidden_states[1:]  # Skip embedding layer
        
        # Weighted sum of layers
        weighted_sum = torch.zeros_like(hidden_states[-1])
        for i, hidden in enumerate(hidden_states):
            weighted_sum += self.layer_weights[i] * hidden
        
        # Mean pooling
        pooled = weighted_sum.mean(dim=1)
        
        # Project to embedding dimension
        embedding = self.embedding_proj(pooled)
        
        # Normalize
        embedding = F.normalize(embedding, p=2, dim=1)
        
        return embedding


def train_architecture_b(train_loader, val_loader, config):
    """Training loop for Architecture B with triplet loss"""
    
    model = WavLMTriplet(config).to(config.DEVICE)
    optimizer = torch.optim.Adam(model.parameters(), lr=config.LEARNING_RATE)
    
    print("\n🚀 Training Architecture B (PEFT + Triplet)")
    print("="*60)
    
    best_loss = float('inf')
    
    for epoch in range(config.EPOCHS):
        # Training
        model.train()
        train_loss = 0
        
        for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{config.EPOCHS} [Train]"):
            anchor = batch['anchor'].to(config.DEVICE)
            positive = batch['positive'].to(config.DEVICE)
            negative = batch['negative'].to(config.DEVICE)
            
            optimizer.zero_grad()
            
            # Get embeddings
            anchor_emb = model(anchor)
            positive_emb = model(positive)
            negative_emb = model(negative)
            
            # Triplet loss
            pos_dist = F.pairwise_distance(anchor_emb, positive_emb, p=2)
            neg_dist = F.pairwise_distance(anchor_emb, negative_emb, p=2)
            
            loss = F.relu(pos_dist - neg_dist + 1.0).mean()
            
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
        
        avg_loss = train_loss / len(train_loader)
        print(f"Epoch {epoch+1}: Train Loss: {avg_loss:.4f}")
        
        # Save best model
        if avg_loss < best_loss:
            best_loss = avg_loss
            torch.save(model.state_dict(), "architecture_b_best.pt")
            print(f"✅ Saved best model with loss: {avg_loss:.4f}")
    
    return model
```

```python
# ============================================
# ARCHITECTURE C: Full Model Fine-tuning
# Unfrozen Top Layers + Triplet Loss
# ============================================

class WavLMFullFineTune(nn.Module):
    """WavLM with top 3 layers unfrozen"""
    
    def __init__(self, config, unfrozen_layers=3):
        super().__init__()
        
        print("🔄 Loading WavLM-large (partially frozen)...")
        self.wavlm = WavLMModel.from_pretrained("microsoft/wavlm-large")
        
        # Freeze all layers first
        for param in self.wavlm.parameters():
            param.requires_grad = False
        
        # Unfreeze top N encoder layers
        if hasattr(self.wavlm, 'encoder') and hasattr(self.wavlm.encoder, 'layers'):
            total_layers = len(self.wavlm.encoder.layers)
            for i in range(total_layers - unfrozen_layers, total_layers):
                for param in self.wavlm.encoder.layers[i].parameters():
                    param.requires_grad = True
            print(f"   Unfroze top {unfrozen_layers} of {total_layers} layers")
        
        # Get hidden size
        wavlm_config = WavLMConfig.from_pretrained("microsoft/wavlm-large")
        hidden_size = wavlm_config.hidden_size
        
        # Trainable components
        self.embedding_proj = nn.Linear(hidden_size, config.EMBEDDING_DIM)
        
        # Count trainable params
        trainable = sum(p.numel() for p in self.parameters() if p.requires_grad)
        total = sum(p.numel() for p in self.parameters())
        
        print(f"✅ Architecture C initialized")
        print(f"   Trainable params: {trainable:,} / {total:,} ({100.*trainable/total:.2f}%)")
    
    def forward(self, audio):
        # Get WavLM outputs
        outputs = self.wavlm(audio, output_hidden_states=True)
        
        # Use last hidden state
        last_hidden = outputs.last_hidden_state
        
        # Mean pooling
        pooled = last_hidden.mean(dim=1)
        
        # Project to embedding dimension
        embedding = self.embedding_proj(pooled)
        embedding = F.normalize(embedding, p=2, dim=1)
        
        return embedding


def train_architecture_c(train_loader, val_loader, config):
    """Training loop for Architecture C"""
    
    model = WavLMFullFineTune(config).to(config.DEVICE)
    optimizer = torch.optim.Adam(model.parameters(), lr=config.LEARNING_RATE * 0.1)  # Lower LR for fine-tuning
    
    print("\n🚀 Training Architecture C (Full Fine-tuning)")
    print("="*60)
    
    best_loss = float('inf')
    
    for epoch in range(config.EPOCHS):
        # Training
        model.train()
        train_loss = 0
        
        for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{config.EPOCHS} [Train]"):
            anchor = batch['anchor'].to(config.DEVICE)
            positive = batch['positive'].to(config.DEVICE)
            negative = batch['negative'].to(config.DEVICE)
            
            optimizer.zero_grad()
            
            # Get embeddings
            anchor_emb = model(anchor)
            positive_emb = model(positive)
            negative_emb = model(negative)
            
            # Triplet loss
            pos_dist = F.pairwise_distance(anchor_emb, positive_emb, p=2)
            neg_dist = F.pairwise_distance(anchor_emb, negative_emb, p=2)
            
            loss = F.relu(pos_dist - neg_dist + 1.0).mean()
            
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
        
        avg_loss = train_loss / len(train_loader)
        print(f"Epoch {epoch+1}: Train Loss: {avg_loss:.4f}")
        
        # Save best model
        if avg_loss < best_loss:
            best_loss = avg_loss
            torch.save(model.state_dict(), "architecture_c_best.pt")
            print(f"✅ Saved best model with loss: {avg_loss:.4f}")
    
    return model
```

```python
# ============================================
# MAIN TRAINING SCRIPT
# ============================================

def main():
    print("\n" + "="*60)
    print("🎯 TRAINING BASELINE ARCHITECTURES")
    print("="*60)
    
    # Create datasets
    print("\n📂 Loading datasets...")
    
    # For Architecture A (classification)
    train_dataset_a = SpeakerDataset(config.SPLIT1_TRAIN, config)
    val_dataset_a = SpeakerDataset(config.SPLIT1_VAL, config)
    
    train_loader_a = DataLoader(
        train_dataset_a, 
        batch_size=config.BATCH_SIZE,
        shuffle=True,
        num_workers=4
    )
    
    val_loader_a = DataLoader(
        val_dataset_a,
        batch_size=config.BATCH_SIZE,
        shuffle=False,
        num_workers=4
    )
    
    # For Architectures B and C (triplet)
    train_dataset_triplet = TripletDataset(config.SPLIT1_TRAIN, config)
    
    train_loader_triplet = DataLoader(
        train_dataset_triplet,
        batch_size=config.BATCH_SIZE,
        shuffle=True,
        num_workers=4
    )
    
    # Train Architecture A
    print("\n" + "="*60)
    model_a = train_architecture_a(train_loader_a, val_loader_a, config)
    
    # Train Architecture B
    print("\n" + "="*60)
    model_b = train_architecture_b(train_loader_triplet, None, config)
    
    # Train Architecture C
    print("\n" + "="*60)
    model_c = train_architecture_c(train_loader_triplet, None, config)
    
    print("\n" + "="*60)
    print("✅ All baseline architectures trained successfully!")
    print("="*60)
    print("\nModels saved:")
    print("  - architecture_a_best.pt")
    print("  - architecture_b_best.pt")
    print("  - architecture_c_best.pt")

if __name__ == "__main__":
    main()
```
